```python
#Sample Example for XOR gate.
import matplotlib.pyplot as plt

x_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]).T  # Shape: (2, 4)
y_train = np.array([[1], [1], [1], [0]]).T  # Shape: (1, 4)

network = NeuralNetwork()
network.add(Dense(2, 3))
network.add(Tanh())
network.add(Dense(3, 1))
network.add(Sigmoid())

# Training
network.train(x_train, y_train, binary_cross_entropy, binary_cross_entropy_prime, epochs=1000, learning_rate=0.1)

# Predict the output for the XOR problem
for x in x_train.T:
    output = network.predict(x.reshape(-1, 1))
    print(f"Input: {x}, Predicted Output: {output.round().flatten()}, True Output: {y_train.T[x_train.T.tolist().index(x.tolist())]}")
```
